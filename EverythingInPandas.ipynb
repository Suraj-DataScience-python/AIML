```python
## Handling missing data

# 1. fillna to fill missing values using different ways
# 2. interpolate to fill missing values 
# 3. dropna to drop row on missing value
```

```python
## Convert day col into TimeStamp
df = pd.read_csv("Sample/sample_read_csv.csv",na_values=['NaN',"Not available"], parse_dates=["EST"])
# df.fillna(0)   ## to replace all na with 0
## fillna with specific columns with specific values
df_nafilled = df.fillna({
    "WindDirDegrees": 0,
    "Events": "No Event"
})
df_nafilled
```

```python
## Fill na values with previous row value
df_nafilled = df.fillna(method="ffill")  ## bfill , axis="columns", limit=1 (to restrict copying)
df_nafilled
```

```python
df_nafilled = df.set_index("EST")
df_nafilled = df_nafilled.interpolate(method='time')  ## slinear, nearest, zero, polynomial, spline
df_nafilled 
```

```python
df_dropna = df.dropna(thresh=1) ## how=all
df_dropna
```

```python
# Replace function : Replace specific col
import numpy as np
df_replace = df.replace({
    "Temperature": -999,
    "CloudCover": -999,
    "Events": None
}, np.NaN)
df_replace
```

```python
## Replace specific row value
df_replace = df.replace({
    -999: np.NaN
})
df_replace

## Using regex: df.replace(['A-Za-z'],np.NaN, regex=True)
```

```python
df_replace = df.replace(["Sunny","Snow","Rain"],[1,2,3])
df_replace
```

```python
## Merge two dataframe

df1 = pd.DataFrame({
    "city":["Delhi", "Mumbai","Chennai","dehradun"],
    "tempreture": [40,25,42,32]
})
```

```python
df2 = pd.DataFrame({
    "city":["Mumbai", "Delhi","Shillong"],
    "humidity": [7,9,5]
})
```

```python
df_merge = pd.merge(df1,df2, on="city")   ## inner join (common element)
df_merge
```

```python
df_merge = pd.merge(df1,df2, on="city", how="outer")   ## outer join (union)
df_merge
```

```python
df_merge = pd.merge(df1,df2, on="city", how="right")   ## left/right join (union)
df_merge
```

```python
df_merge = pd.merge(df1,df2, on="city", how="outer", indicator=True)   ## To know from where the came
df_merge
```
